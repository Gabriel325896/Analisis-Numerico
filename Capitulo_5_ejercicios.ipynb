# Problema 3

imagen.png

## (a)

\begin{equation}
\begin{aligned}
\min \quad & \sum^{24}_{t=1} a_{t} x_{t}^{2} + b_{t} x_{t} + c_t\\
\textrm{s.t.} \quad & \sum^{24}_{t=1} x_{t} = E_{Tot} \\
  &x_{t}\geq0    \\
\end{aligned}
\end{equation}

## (b)

Se re escribe el problema como:

\begin{equation}
\begin{aligned}
\min \quad & \mathbf{x}^{T}\mathbf{A} \mathbf{x} + \mathbf{B}\mathbf{x} + \mathbf{C}\\
\textrm{s.t.} \quad & \mathbf{A_r}\mathbf{x} = E_{Tot} \\
  &x_{t}\geq0    \\
\end{aligned}
\end{equation}

Este es un problema cuadratico. El langrangiano seria

$$L(\mathbf{x},\mathbf{v}) = \mathbf{x}^{T}\mathbf{A} \mathbf{x} + \mathbf{B}\mathbf{x} + \mathbf{C} + \mathbf{v}^{T}(\mathbf{A_r}\mathbf{x} - E_{Tot})$$

El gradiente de este seria:

$$\nabla_{x}L(\mathbf{x},\mathbf{v}) = \mathbf{A}\mathbf{x} + \mathbf{B} + \mathbf{A_{r}}^{T} \mathbf{v}$$

Pr tanto las KKT condiciones se puede escribir así

$$\begin{bmatrix}
 \mathbf{A} & \mathbf{A_r}^{T}  \\
 \mathbf{A_r}& 0
\end{bmatrix}\begin{bmatrix}
 \mathbf{x}   \\
 \mathbf{v}
\end{bmatrix}  = \begin{bmatrix}
 \mathbf{-B}   \\
E_{Tot}
\end{bmatrix}$$
