# Problema 5:

image.png

```python
Ap5_a = np.array([[1,-3],
              [4,10]])
print('Matriz original = ')
print(Ap5_a)
evalsp5_a, evecsp5_a = np.linalg.eig(Ap5_a)
Diagp5_a = evecsp5_a@np.diag(evalsp5_a)@np.linalg.inv(evecsp5_a)
print('Matriz diagonalizada = ')
print(Diagp5_a)
```

```python
Ap5_b = np.array([[4, 3, 6],
                  [3, -8, 15],
                  [6, 15, 10]])
print('Matriz original = ')
print(Ap5_b)
evalsp5_b, evecsp5_b = np.linalg.eig(Ap5_b)
Diagp5_b = evecsp5_b@np.diag(evalsp5_b)@np.linalg.inv(evecsp5_b)
print('Matriz diagonalizada = ')
print(Diagp5_b)
```

# Problema 9:

image.png

```python
feature_x = np.arange(-2, 2+0.2, 0.2)
feature_y = np.arange(-2, 2+0.2, 0.2)

x, y = np.meshgrid(feature_x, feature_y)
z = ((x**4) + 2*(x**3)*y - 6*(x**2)*(y**2) + (y**4))/((x**4) + (y**4) + 1)
u = -(2*x*((x**5)*y - 6*(x**4)*(y**2) - 2*(x**2) - 3*x*((y**5) + y) + 6*((y**6) + (y**2))))/ ( ((x**4) + (y**4) + 1)**2)
v =  (2*((x**7) - 6*(x**6)*y + (x**3)*(1 - 3*(y**4)) + 6*(x**2)*y*((y**4) - 1) + 2*(y**3)))/((x**4 + (y**4) + 1)**2)

# Normalize all gradients to focus on the direction not the magnitude
norm = np.linalg.norm(np.array((u, v)), axis=0)
u = u / norm
v = v / norm

fig, ax = plt.subplots(1, 1)
ax.quiver(x, y, u, v, units='xy', scale=6.2, color='gray')
ax.contour(x, y, z, 10, cmap='jet')
plt.show()
```

```python
A = np.array([[1,0,2,2],
              [-2,42,24,4],
              [7,-21,0,-5]])

rank_a = np.linalg.matrix_rank(A)
At = np.transpose(A)
rank_b = np.linalg.matrix_rank(At)

if rank_a == rank_b:
  print('Los rangos son iguales.')
else:
  print('Los rangos no son iguales.')
```
